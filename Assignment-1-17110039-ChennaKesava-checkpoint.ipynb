```python
import csv
import matplotlib.pyplot as plt
import collections
import re
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
nltk.download('punkt')
```

# Tokens,Types & TTR

```python
all_tokens=[]
tok_sor=[]
with open('tweets.csv','r',encoding="utf8") as csv_file:
    reader = csv.reader(csv_file)
    for i in reader:
        tokens=nltk.word_tokenize(i[0])
        for j in tokens:
            all_tokens.append(j)
    fdist = nltk.FreqDist(all_tokens)
    
    tok_sor = sorted(fdist.items(),key=lambda k:k[1])
    print("No. of types :",len(tok_sor))
    print("No. of tokens :",len(all_tokens))
    print("TTR :",len(tok_sor)/len(all_tokens))
# for k in tok_sor:
#     if(k[1]>=26):

        
#         synonyms = []

#         for syn in wordnet.synsets(k[0]):
#             for l in syn.lemmas():
#                 synonyms.append(l.name())


#         if(len(set(synonyms))>=26):
#             print(k,len(set(synonyms)))
                

    
```

# Heaps Law

```python

x=[]
y=[]
for i in range(0,len(all_tokens),960):
    y.append(len(nltk.FreqDist(all_tokens[0:i])))
    x.append(len(all_tokens[0:i]))
    

    
plt.plot(x,y,label = "according to data")
plt.xlabel('No. of Tokens')
plt.ylabel('Vocabulary')
plt.title("Heaps Law")
```

# Zipf's law of length

```python
# print(tok_sor)
B={} #key=length value=frequency of that length

for i in all_tokens:
    if(len(i) not in B):
        B[len(i)]=1
    else:
        B[len(i)]+=1


srtd = collections.OrderedDict(sorted(B.items()))
# print(srtd)
x=list(srtd.keys())
y=list(srtd.values())

plt.plot(x,y)
    
plt.xlabel('Frequency')
plt.ylabel('Length')
plt.title("Zipf's law of lenght")
```

# Zipf's law of meaning


```python
zz=[]
for i in all_tokens:
    f=re.findall(r"[A-Za-z]+",i)
    zz.append(f)
# print(zz)


A={}
MD={}
mean_words = ["problem","Humor","support","balance","support","donate","Free","Good","black","thought","thinking","seat","give","block"]
for i in mean_words:
    for j in tok_sor:
        if(j[0]==i):
            x.append(j[1])


sd=[]            
for i in mean_words:            
    synonyms = []

    for syn in wordnet.synsets(i):
        for l in syn.lemmas():
            synonyms.append(l.name())

    for j in tok_sor:
        if(j[0]==i):

            A[j[1]]=len(set(synonyms))
            MD[j[1]]=[len(set(synonyms)),i]
            sd.append(i)

# print(MD)            

srtdmen = collections.OrderedDict(sorted(A.items()))
# print(srtdmen)
x=list(srtdmen.keys())
y=list(srtdmen.values())
# print(x)
# print(y)

plt.plot(x,y)
plt.xlabel('Frequency')
plt.ylabel('Number of Meanings')
plt.title("Zipf's law of meaning")
```
